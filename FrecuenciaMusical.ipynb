```python
import matplotlib.pyplot as plt
import numpy as np
from math import sqrt,log
```

```python
t=np.round(np.arange(0,3.01,0.01),2)
A=1
def x(t,A,f):
  return A*np.sin(f*t)
```

```python
notas=["Do","Re","Mi","Fa","Sol","La","Si"]
```

```python
hz=440
octavas=11
```

```python
fig,ax=plt.subplots(octavas,12)
fig.set_size_inches(20,20)
for i in range(12*octavas):
  h=hz*pow(2**(1/12),i-69)
  h=round(h,3)
  #print(h,round(12*log(h/hz,2))+69)
  #print(notas[(i+5-(i>1)*7)],str(h)+"Hz")
  ax[i//12,i%12].plot(t,x(t,A,h),color=[(i%2)/2+(i>5)*0.5,(i%2==0)*1,(i%4)/4])
```

```python
nota_music="La2"
```

```python
if nota_music[:2] in notas:
  ind=notas.index(nota_music[:2])
  if nota_music[2:].isnumeric():
    midi=ind*2-(ind>=3)*1+(int(nota_music[2:])+1)*12
  if nota_music[2]=="b":
    if len(nota_music)>3:
      if nota_music[3:].isnumeric():
        midi=ind*2+(ind<3)*1+(int(nota_music[3:])+1)*12
if nota_music[:3]=="Sol":
  ind=notas.index(nota_music[:3])
  if nota_music[3:].isnumeric():
    midi=ind*2-(ind>=3)*1+(int(nota_music[3:])+1)*12
  if nota_music[3]=="b":
    if len(nota_music)>4:
      if nota_music[4:].isnumeric():
        midi=ind*2+(ind<3)*1+(int(nota_music[4:])+1)*12

fn=round(hz*pow(2**(1/12),midi-69),3)
longitud_onda=round(343/fn,4)
print("f:",str(fn)+"Hz")
print(":",str(longitud_onda)+"m")
print("Nota midi:",(midi<128)*midi)
plt.plot(t,x(t,A,fn))
```
